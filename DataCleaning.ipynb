### Initial Data Cleaning

```python
# Extract country information and clean player names
def extract_country_and_clean_name(player_name):
    parts = player_name.split('(')
    name = parts[0].strip()
    country = parts[-1].replace(')', '').strip()
    return pd.Series([name, country])

df[['Player Name', 'Country']] = df['Player Name'].apply(extract_country_and_clean_name)
```

```python
df.head()
```

```python
# Define a function to convert the date format
def convert_date_format(date_str):
    try:        
        # Extract the date part if there are additional details
        month = date_str.split(',')[0]
        year = date_str.split(',')[1]
        date_part = month + " " + year
        # Try to parse the date in different possible formats
        date_obj = pd.to_datetime(date_part, errors='coerce')
        if pd.notnull(date_obj):
            return date_obj.strftime('%m-%d-%Y')
        else:
            return pd.to_datetime(date_str, errors='coerce')
        
    except Exception as e:
        print(f"Error parsing date: {date_str}, Error: {e}")
    return date_str

# Apply the function to the Born column
df['Born'] = df['Born'].apply(convert_date_format)
```

```python
df.head()
```

```python
# Clean the HS_ODI, HS_T20, and HS_Test columns
df['HS_ODI'] = df['HS_ODI'].str.replace('*', '', regex=False)
df['HS_ODI'] = df['HS_ODI'].replace('-', '0')

df['HS_T20'] = df['HS_T20'].str.replace('*', '', regex=False)
df['HS_T20'] = df['HS_T20'].replace('-', '0')

df['HS_Test'] = df['HS_Test'].str.replace('*', '', regex=False)
df['HS_Test'] = df['HS_Test'].replace('-', '0')
```

```python
# Create the Current Status column
def determine_current_status(row):
    if 'Currently Playing' in [row['Current Status_ODI_batting'], row['Current Status_T20_batting'], row['Current Status_Test_batting']]:
        return 'Currently Playing'
    else:
        return 'Retired'

df['Current Status'] = df.apply(determine_current_status, axis=1)
```

```python
# Drop additional Span Columns
df['SpanStart_ODI'] = df['Span Start_ODI_batting']
df['SpanEnd_ODI'] = df['Span End_ODI_batting']
df['SpanStart_Test'] = df['Span Start_Test_batting']
df['SpanEnd_Test'] = df['Span End_Test_batting']
df['SpanStart_T20'] = df['Span Start_T20_batting']
df['SpanEnd_T20'] = df['Span End_T20_batting']
```

```python
# Define a function to convert age to decimal format
def convert_age_to_decimal(age_str):
    try:
        if pd.isnull(age_str):
            return None
    
        years = 0
        days = 0 
        parts = age_str.split('y')
        
        if len(parts[0]) >=1: 
            years = float(parts[0])

        if len(parts[1]) >= 1: 
            days = float(parts[1][:-1]) # Remove the trailing 'd'
         
        return years + days / 365
    except Exception as e:
        print(f"Error parsing age: {age_str}, Error: {e}")
        return None

df['Age'] = df['Age'].apply(convert_age_to_decimal)
```

```python
# Drop the old Current Status columns
df = df.drop(columns=['Span End_T20_batting','Span Start_T20_batting','Span End_Test_batting','Span Start_Test_batting','Span End_ODI_batting','Span Start_ODI_batting', 'Current Status_ODI_batting', 'Current Status_T20_batting', 'Current Status_Test_batting', 'Current Status_ODI_bowling', 'Current Status_T20_bowling', 'Current Status_Test_bowling', 'Height', 'SR_Test_batting'])
```
